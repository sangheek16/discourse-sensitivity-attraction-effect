```python
import pandas as pd
```

```python
comparisons = pd.read_csv('clause_comparison_table.tsv', sep='\t')
```

## C1

```python
subset = comparisons[comparisons['exp'] == 'exp1']
subset = subset[subset['Grammaticality'] == 'ungram']
subset = subset[subset['overlaps'] == 'no']
ungram_models_exp1 = set(subset['model'].unique())
```

```python
for model in ungram_models_exp1:
    print(model)
```

## C2 Weak

```python
differences = pd.read_csv('by_clause_difference_table.tsv', sep='\t')
differences = differences[differences['exp'] == 'exp2']
differences = differences[differences['Grammaticality'] == 'ungram']
differences = differences[differences['ContainsZero'] == 'no']
differences = differences.groupby(['model']).count().reset_index()[['model', 'exp']]
differences = differences[differences['exp'] == 2]
C2 = set(differences['model'].tolist())
for model in differences['model'].tolist():
    print(model)
print(len(differences['model'].tolist()))
```

## C2 Strong

```python
subset = comparisons[comparisons['exp'] == 'exp2']
subset = subset[subset['Grammaticality'] == 'ungram']
subset = subset[subset['overlaps'] == 'yes']
ungram_models_exp2 = set(subset['model'].unique())
ungram_models_exp2 = list(ungram_models_exp2)
ungram_models_exp2.sort()
for model in ungram_models_exp2: 
    if model in C2:
        print(model)
```
